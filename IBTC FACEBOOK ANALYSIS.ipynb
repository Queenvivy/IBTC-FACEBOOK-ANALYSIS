**Changing all numeric columns to integers**

```python
# Select all numeric columns using select_dtypes
numeric_columns = IBTC.select_dtypes(include='number')

# Replace NaN values with 0 in numeric columns
IBTC[numeric_columns.columns] = numeric_columns.fillna(0)


# Define a function to remove commas and convert to integers
def remove_commas_and_convert(value):
    try:
        # Convert to integer if possible
        return int(str(value).replace(',', ''))
    except (ValueError, TypeError):
        # If conversion is not possible, return as is
        return value

# Apply the function to all columns in the DataFrame
IBTC = IBTC.applymap(remove_commas_and_convert)

```
